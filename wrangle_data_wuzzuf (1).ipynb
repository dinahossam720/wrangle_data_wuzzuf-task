# cleaning

```python
df_jobs_clean =df_jobs.copy()
df_apps_clean =df_apps.copy()

```

```python
#define:convert post_date column in df_jobs_clean  to datatime dtype 


#Code:
df_jobs_clean['post_date'] = pd.to_datetime( (df_jobs_clean['post_date'].str.slice(start=0, stop=-6) ), format = "%Y-%m-%d %H:%M:%S")

#Test
df_jobs_clean.info()
```

```python
#define:convert app_date column in df_apps_clean  to datatime dtype 


#Code:
df_apps_clean['app_date'] = pd.to_datetime( (df_apps_clean['app_date'].str.slice(start=0, stop=-6) ), format = "%Y-%m-%d %H:%M:%S")

#Test
df_apps_clean.info()
```
